### Препроцессинг данных

```python
# Удаляем вспомогательные при генерации столбцы, чтобы не было ликов при обучении
customers.drop(columns=['available_terminals', 'nb_terminals'], inplace=True)
transactions.drop(columns=['TX_FRAUD_SCENARIO'], inplace=True)
```

```python
# Удаляем мусорные признаки
transactions.drop(columns=['TX_TIME_SECONDS', 'TX_TIME_DAYS'], inplace=True)
```

```python
# Сводим все в одну таблицу
result = pd.merge(transactions, customers, on=["CUSTOMER_ID", "CUSTOMER_ID"])
result = pd.merge(result, terminals, on=["TERMINAL_ID", "TERMINAL_ID"])
result.sort_values(by='TX_DATETIME', inplace=True)
```

```python
# Работаем с временными признаками
result['TX_DATETIME'] = pd.to_datetime(result['TX_DATETIME'])
result['day_of_week'] = result['TX_DATETIME'].dt.dayofweek
result['hour'] = result['TX_DATETIME'].dt.hour
result.drop(columns=['TX_DATETIME'], inplace=True)
```
